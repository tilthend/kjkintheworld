```python
pop_seoul.drop([0], inplace=True)
```

```python
pop_seoul['시군구명'].unique()
```

```python
pop_seoul = pop_seoul.loc[:,['시군구명','총생활인구수','내국인생활인구수','자치구간이동인구수']]
```

```python
pop_seoul
```

```python
pop_seoul['외국인생활인구수'] = pop_seoul['총생활인구수']-pop_seoul['내국인생활인구수']
```

```python
pop_seoul
```

```python
pop_seoul[pop_seoul['시군구명'].isnull()]
```

```python
pop_seoul['외국인생활인구수비율'] = pop_seoul['외국인생활인구수']/pop_seoul['총생활인구수']*100
pop_seoul['내국인생활인구수비율'] = pop_seoul['내국인생활인구수']/pop_seoul['총생활인구수']*100
pop_seoul['자치구간이동인구수비율'] = pop_seoul['자치구간이동인구수']/pop_seoul['총생활인구수']*100
```
