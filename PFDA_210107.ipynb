#### 계단 오르내리기 예제...

```python
import random
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

#### xrange는 range로 대체...

```python
nsteps = 1000
draws = np.random.randint(0, 2, size = nsteps)
```

```python
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

```python
(np.abs(walk) >= 10).argmax()
```

```python
nwalks = 5000
```

```python
nsteps = 1000
```

```python
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
```

```python
steps = np.where(draws > 0, 1, -1)
```

```python
walks = steps.cumsum(1)
```

```python
walks
```

```python
walks.max()
```

```python
walks.min()
```

```python
hits30 = (np.abs(walks) >= 30).any(1)
```

```python
hits30
```

```python
hits30.sum()
```

```python
crossing_times = (np.abs(walks[hits30]) >=  30).argmax(1)
```

```python
crossing_times.mean()
```
