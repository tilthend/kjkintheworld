```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding = 'UTF8')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[3] != '' :
        if row[1].split('-')[1] == '08' :
            result.append(float(row[3]))
plt.plot(result, 'red', label = 'temp')
plt.legend(loc=4)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding = 'UTF8')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[3] != '' :
        if row[1].split('-')[1] == '08' and row[1].split('-')[2] == '01' :
            result.append(float(row[3]))
plt.plot(result, 'red', label = 'temp')
plt.legend(loc=4)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding = 'UTF8')
data = csv.reader(f)
next(data)

high = []
low = []

for row in data :
    if row[3] != '' and row[5] !='' :
        if 1986 <= int(row[1].split('-')[0]) :
            if row[1].split('-')[1] == '08' and row[1].split('-')[2] == '01' :
                high.append(float(row[3]))
                low.append(float(row[5]))
            
plt.plot(high, 'red', label = 'high')
plt.plot(low, 'black', label = 'low')

plt.legend(loc=4)
plt.show()
```
