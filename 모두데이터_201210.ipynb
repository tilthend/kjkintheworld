```python
import csv
f = open('age11.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하라 : ')

for row in data :
    if name in row[0] :
        for i in row[3:] :
            result.append(int(i.replace(',','')))
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(name+' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
import csv
f = open('age.csv')
data = csv.reader(f)

m = []
f = []

result = []
for row in data :
    if '방이2동' in row[0] :
        for i in range(0,101) :
            m.append(int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()

import matplotlib.pyplot as plt
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import csv
f = open('age.csv')
data = csv.reader(f)
m=[]
f=[]
for row in data :
    if '방이2동' in row[0] :
        for i in row[106:207] :
            m.append(int(i))
        for i in row[209:] :
            f.append(-int(i))
            
import matplotlib.pyplot as plt
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import csv
f = open('age.csv')
data = csv.reader(f)
m=[]
f=[]
for row in data :
    if '방이2동' in row[0] :
        for i in row[106:207] :
            m.append(int(i))
        for i in row[209:] :
            f.append(-int(i))
            
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend(loc=4)
plt.show()
```
