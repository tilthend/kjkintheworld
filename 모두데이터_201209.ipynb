```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding = 'UTF8')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[5] != '':
        result.append(float(row[5]))
plt.hist(result, bins=100, color='r') #히스토그램
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding = 'UTF8')
data = csv.reader(f)
next(data)
aug = [] #리스트 생성

for row in data :
    month = row[1].split('-')[1] 
    if row[5] != '' :
        if month == '08':
            aug.append(float(row[5]))
            
plt.hist(aug, bins=100, color='r') #히스토그램
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding = 'UTF8')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data :
    month = row[1].split('-')[1]
    if row[5] != '' :
        if month == '08':
            aug.append(float(row[5]))
        if month == '01':
            jan.append(float(row[5]))

            
plt.hist(aug, bins=100, color='r', label = 'Aug')
plt.hist(jan, bins=100, color='b', label = 'Jan')

plt.legend(loc=1)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding = 'UTF8')
data = csv.reader(f)
next(data)

month = [],[],[],[],[],[],[],[],[],[],[],[]

for row in data :
    if row[5] != '':
        month[int(row[1].split('-')[1])-1].append(float(row[5]))
            
plt.boxplot(month)

plt.show()
```
