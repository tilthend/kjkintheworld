```python
import pandas as pd
```

```python
CCTV_Seoul = pd.read_csv('C:/Users/woowahan/Documents/data/Python/datamassage/cctv.csv', encoding = 'utf8')
```

```python
pop_seoul=pd.read_excel('C:/Users/woowahan/Documents/data/Python/datamassage/pop_seoul.xlsx', enconding='utf-8' )
```

```python
pop_seoul.drop([0], inplace=True)
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0] : '시군구명'}, inplace=True)
```

```python
data_result = pd.merge(CCTV_Seoul, pop_seoul, on='시군구명')
```

```python
data_result.head()
```

```python
data_result.set_index('시군구명', inplace=True)
```

```python
data_result.head()
```

```python
del data_result['기준일ID']
del data_result['일최대인구수']
del data_result['일최소인구수']
del data_result['주간인구수(09~18)']
del data_result['시군구코드']
del data_result['야간인구수(19~08)']
```

```python
data_result.head()
```

```python
del data_result['2013년']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
del data_result['2017년']
del data_result['2018년']
```

```python
data_result.head()
```

```python
del data_result['2012년']
del data_result['2011년 이전']
```

```python
data_result.head()
```

```python
data_result['외국인생활인구수'] = data_result['총생활인구수'] - data_result['내국인생활인구수']
```

```python
data_result['외국인생활인구수비율'] = data_result['외국인생활인구수'] / data_result['총생활인구수'] * 100
data_result['내국인생활인구수비율'] = data_result['내국인생활인구수'] / data_result['총생활인구수'] * 100
```
