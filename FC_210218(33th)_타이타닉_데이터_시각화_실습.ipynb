```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# 한글 폰트 적용
plt.rc('font', family='NanumBarunGothic') 
# 캔버스 사이즈 적용
plt.rcParams["figure.figsize"] = (12, 9)
```

[공식 도큐먼트](https://seaborn.pydata.org/api.html)

```
titanic = sns.load_dataset('titanic')
titanic
```

* survived: 생존여부
* pclass: 좌석등급
* sex: 성별
* age: 나이
* sibsp: 형제자매 + 배우자 숫자
* parch: 부모자식 숫자
* fare: 요금
* embarked: 탑승 항구
* class: 좌석등급 (영문)
* who: 사람 구분
* deck: 데크
* embark_town: 탑승 항구 (영문)
* alive: 생존여부 (영문)
* alone: 혼자인지 여부

```
tips = sns.load_dataset('tips')
tips
```

* total_bill: 총 합계 요금표
* tip: 팁
* sex: 성별
* smoker: 흡연자 여부
* day: 요일
* time: 식사 시간
* size: 식사 인원

```
sns.set(style='dark')
```

```
# 세로 그래프

sns.countplot(x="class", hue="who", data=titanic)
plt.show()
```

```
# 가로로 그리기

sns.countplot(y="class", hue="who", data=titanic)
plt.show()
```

```
#색상 팔레트 설정 -> palletter

sns.countplot(x="class", hue="who", palette='crest_r', data=titanic)
plt.show()
```

matplotlib의 `hist` 그래프와 `kdeplot`을 통합한 그래프

**분포**와 **밀도**를 확인할 수 있음

[distplot 도큐먼트](https://seaborn.pydata.org/generated/seaborn.distplot.html?highlight=distplot#seaborn.distplot)

```
# 샘플데이터 생성
x = np.random.randn(100)
x
```

```
sns.distplot(x)
plt.show()
```

data type이 series인 경우

```
x = pd.Series(x, name="x variable")
x
```

```
sns.distplot(x)
plt.show()
```

`rug`는 `rugplot`이라고도 불리우며, 데이터 위치를 x축 위에 **작은 선분(rug)으로 나타내어 데이터들의 위치 및 분포**를 보여줌


```
sns.distplot(x, rug=True, hist=False, kde=True)
plt.show()
```

kde(kernel density)

히스토그램보다 부드러운 형태의 분포 곡선 표현

```
sns.distplot(x, rug=False, hist=False, kde=True)
plt.show()
```

```
 sns.distplot(x, vertical=True)
```

```
sns.distplot(x, color="g")
plt.show()
```

히트맵

[heatmap 도큐먼트](https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap)

[pairplot 도큐먼트](https://seaborn.pydata.org/generated/seaborn.pairplot.html?highlight=pairplot#seaborn.pairplot)

pairplot은 그리도(grid) 형태로 각 **집합의 조합에 대해 히스토그램과 분포도**를 그립니다.

또한, 숫자형 column에 대해서만 그려줍니다.

```
tips.head()
```

```
sns.pairplot(tips)
plt.show()
```

```
sns.pairplot(tips, hue='size')
plt.show()
```

```
sns.pairplot(tips, hue='size', palette="rainbow")
plt.show()
```

```
sns.pairplot(tips, hue='size', palette="rainbow", height=5,)
plt.show()
```
