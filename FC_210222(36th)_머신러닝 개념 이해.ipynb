데이터 기반으로 패턴을 학습, 결과를 추론해 보자!

 - 지도학습: regression(numeric value), Classification(categorical value)
 - 비지도학습: Clustering(groupinb), Dimentionality Reduction

 머신러닝은...
 - 복잡한 패턴 인지
 - 적절한 알고리즘, 양질 데이터로 성능 좋음
 - 도메일 지식 부족해도 분석 가능

 문제점은...
 - Garbage in, Garbage out
 - 과적합 오류(일반화 or 데이터 당양성 필요)
 - 풍부한 데이터 필요

 데이터, 예측해야 할 값에 맞는 적절한 알고리즘 필요하다!
 



*가설함수(Hypothesis)*
- 가설함수는 모델을 예측하는 근간이 되는 역할
- H(x)로 표현
- H(x) = W*X+b
가설함수를 적용해 나온 Y 값을 Y predict라고 부른다면

손실 = Y Predict - Y 즉 예측값과의 차이(오차라고 함), LOSS

손실함수 = W*X+b - Y

가장 좋은 알고리즘은 손실함수의 값이 0으로 수렵해야 한다.

손실함수(Loss Function)

= W*X + B - Y

=(W*X + b - Y)^2 의 전체 합

데이터 개수가 많아지면 손실이 커지므로 전체 손실의 평균 구한다.

∑(W*X+b - Y)^2 / N
