# 타이타닉 데이터 실습

```
import numpy as np
import pandas as pd
```

```
train = pd.read_csv('https://bit.ly/fc-ml-titanic')
```

```
train.head()
```

* PassengerId: 승객 아이디
* Survived: 생존 여부, 1: 생존, 0: 사망
* Pclass: 등급
* Name: 성함
* Sex: 성별
* Age: 나이
* SibSp: 형제, 자매, 배우자 수
* Parch: 부모, 자식 수
* Ticket: 티켓번호
* Fare: 요즘
* Cabin: 좌석번호
* Embarked: 탑승 항구

train과 valid set 나눠줌.

피쳐랑 레이블 먼저 정의하기

```
feature = [
    'Pclass', 'Sex', 'Age', 'Fare'
]
```

```
label = [
    'Survived'
]
```

```
train[feature].head()
```

```
train[label].head()
```

```
from sklearn.model_selection import train_test_split
```

* **test_size**: validation set에 할당할 비율 (20% -> 0.2)
* **shuffle**: 셔플 옵션 (기본 True)
* **random_state**: 랜덤 시드값

```
x_train, x_valid, y_train, y_valid = train_test_split(train[feature], train[label], test_size=0.2, shuffle=True, random_state=30)
```
